## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x*y*z

Wolfram7.JPG

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y/z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))

w=N(1/(pi + I), 20)
print('w=',w)
```

```
# Ejemplo en Sympy
from sympy import *
x1, x2, x3, x4, y1, y2, y3, y4, z1, z2, z3, z4 = symbols("x1 x2 x3 x4 y1 y2 y3 y4 z1 z2 z3 z4", real=True)  
x = x1 + I*x2 + I*x3 + I*x4
y = y1 + I*y2 + I*y3 + I*y4
z = z1 + I*z2 + I*z3 + I*z4

print(x*y+z)
print(expand(x*y+z))
print(expand((x*y)+z))
print(expand(x*(y+z)))

w=N(1/(pi + I), 20)
print('w=',w)
```
